## Adjective list constitution

```python
import spacy
import os
from tqdm import tqdm

# Load the English language model
nlp = spacy.load('en_core_web_sm')
adj_list = {}

import_dir = "RapGeniusScrapper/Results"
for dir_name in tqdm(os.listdir(import_dir)):
    for file_name in os.listdir(os.path.join(import_dir, dir_name)):
        with open(os.path.join(import_dir, dir_name, file_name), "r", encoding="utf-8") as f:
            text = f.read()
        doc = nlp(text)
        for token in doc:
            if( token.pos_ == "ADJ"):
                if token.text in adj_list.keys():
                    adj_list[token.text] += 1
                else:
                    adj_list[token.text] = 1
```

```python
adj_list = sorted(adj_list.items(), key=lambda x: x[1], reverse=True)
```
